```python
import re
import pandas as pd
```

```python
f = open('sample/WhatsApp Chat with Btech Cs.txt', 'r', encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern= '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s\D[am|pm]\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
patterns = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s\D[am|pm]'
dates = re.findall(patterns, data)
dates
```

```python
# convert from 12 hour format to 24 hour format
from dateutil import parser
date = []
for i in dates:
    x = parser.parse(i)
    date.append(x)
print(date)
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': date})
# convert message_date type
df['message_date'] = pd.to_datetime(df['message_date'])

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```
